```python
import yfinance as yf
import pandas as pd
import requests
from bs4 import BeautifulSoup
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm"
response = requests.get(url)
html_data = response.text
```

```python
soup = BeautifulSoup(html_data, "html.parser")
```

```python
# Find all tables
tables = soup.find_all('table')
relevant_table = None

# Identifying the relevant table
for table in tables:
        if 'Tesla Quarterly Revenue' in table.text:
            relevant_table = table
            break

if relevant_table is None:
    raise ValueError("Couldn't find a table with 'Tesla Quarterly Revenue'")

# Initializing a dataframe with columns Date and Revenue
tesla_revenue = pd.DataFrame(columns = ['Date', 'Revenue'])

# Loop through row to find the relevant table and extract data from 1st and 2nd columns
for row in relevant_table.find_all('tr'):
    cols = row.find_all('td')
    if len(cols) >= 2:
        date = cols[0].text.strip()
        revenue = cols[1].text.strip()
        tesla_revenue = pd.concat([tesla_revenue, pd.DataFrame([[date, revenue]], columns = ["Date", "Revenue"])], ignore_index=True) # adding rows to dataframe
```

```python
tesla_revenue["Revenue"] = tesla_revenue['Revenue'].str.replace(',|\$',"", regex=True)
```

```python
tesla_revenue.dropna(inplace=True)
tesla_revenue = tesla_revenue[tesla_revenue['Revenue'] != ""]
```
