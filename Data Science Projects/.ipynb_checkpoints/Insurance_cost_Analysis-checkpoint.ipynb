# Insurance Cost Analysis

### Importing required libraries into the notebook

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import cross_val_score, train_test_split
```

### Loading and reading the insurance dataset 

```python
filename = "insurance.csv"
```

```python
df = pd.read_csv(filename)
df.head(10)
```

### Add Headers to the columns

```python
headers = ['age', 'gender', 'bmi', 'no_of_children', 'smoker', 'region', 'charges']
df.columns = headers
df.head(10)
```

### Replace the missing values with NaN

```python
df.replace('?', np.nan, inplace= True)
df.head(10)
```

```python
df.info()
```

## Data Wrangling
### Handling missing data

```python
mean_age = df['age'].astype('float').mean(axis=0)
df["age"].replace(np.nan, mean_age, inplace = True)

is_smoker = df['smoker'].value_counts().idxmax()
df["smoker"].replace(np.nan, is_smoker, inplace=True)

df[["age","smoker"]] = df[["age","smoker"]].astype("int")
df.info()
```

```python
df[["charges"]] = np.round(df[["charges"]],2)
df.head()
```

```python
df.info()
```

## Exploratory Data Analysis

### Regrssion plot for charges wrt bmi

```python
sns.regplot(x="bmi", y="charges", data=df, line_kws={"color": "red"})
plt.ylim(0,)
```

### Box plot for charges wrt smoker

```python
sns.boxplot(x = 'smoker', y = 'charges', data = df)
```

```python
df.corr()
```

## Model Development

```python
lm = LinearRegression()
X = df[['smoker']]
Y = df[['charges']]
lm.fit(X,Y)
lm.score(X,Y)
```

### Fitting a LR model used to predict charges with all other attributes of dataset

```python
lm = LinearRegression()
Z = df[['smoker', 'age', 'gender', 'bmi', 'no_of_children', 'region']]

lm.fit(Z,Y)
lm.score(Z,Y)
```

```python
Input=[('scale',StandardScaler()), ('polynomial', PolynomialFeatures(include_bias=False)), ('model', LinearRegression())]
pipe=Pipeline(Input)
Z = Z.astype(float)
pipe.fit(Z,Y)
ypipe=pipe.predict(Z)
print(r2_score(Y,ypipe))
```

## Model Refinement

```python
x_train, x_test, y_train, y_test = train_test_split(Z,Y,test_size = 0.2, random_state = 1)
# x_train, x_test, y_train, y_test 
```

```python
RidgeModel=Ridge(alpha=0.1)
RidgeModel.fit(x_train, y_train)
yhat = RidgeModel.predict(x_test)
print(r2_score(y_test,yhat))
```

```python
pr = PolynomialFeatures(degree=2)
x_train_pr = pr.fit_transform(x_train)
x_test_pr = pr.fit_transform(x_test)
RidgeModel.fit(x_train_pr, y_train)
y_hat = RidgeModel.predict(x_test_pr)
print(r2_score(y_test,y_hat))
```
